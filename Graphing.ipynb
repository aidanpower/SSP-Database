# Graphing Citation Networks

This notebook contains the code necessary for graphing the citation network as it grows. 

```python
# Modules for Matplotlib
import pandas as pd
import metaknowledge as mk
import networkx as nx
import matplotlib.pyplot as plt
import community

from collections import defaultdict
from matplotlib.offsetbox import AnchoredText

# Modules for Holoviews
import holoviews as hv
import bokeh

from holoviews.operation.datashader import bundle_graph, datashade

hv.extension('bokeh')
```

```python
RC_ALL = mk.RecordCollection('SSP Citations/')
RC_2014 = mk.RecordCollection('SSP Citations/SSP 2014.txt')
RC_2015 = mk.RecordCollection('SSP Citations/SSP 2015.txt')
RC_2016 = mk.RecordCollection('SSP Citations/SSP 2016.txt')
RC_2017 = mk.RecordCollection('SSP Citations/SSP 2017.txt')
RC_2018 = mk.RecordCollection('SSP Citations/SSP 2018.txt')
```

```python
# merge joins two RecordCollections into one
# merge: RecordCollection RecordCollection --> RecordCollection
def merge(RC1, RC2):
    NewRC = mk.RecordCollection()
    for R in RC1:
        NewRC.add(R)
    for R in RC2:
        NewRC.add(R)
    return NewRC

# sort_communities takes a dictionary where the keys are the nodes' labels and the values are their respective communities and returns a dictionary
# where each community is the key and the value is the list of nodes in that particular community.
# sort_communities: dictof(Str:Int) --> dictof(Int: listof(Str))
def sort_communities(partition):
    communities = {}
    for key, value in sorted(partition.items()):
        communities.setdefault(value, []).append(key)

    return communities
```

# Plotting Matplotlib Graphs

The following code creates graphs using Matplolib and can be used to show community detection, as well as fixed place labels.

### Citation Network 2014 

```python
RC = RC_2014

G = RC.networkCitation(coreOnly=True)
mk.dropEdges(G, minWeight = 1, dropSelfLoops = True)
G.remove_nodes_from(list(nx.isolates(G)))

G1 = G.to_undirected()
G1_giant = max(nx.connected_component_subgraphs(G1), key=len)
partition = community.best_partition(G1_giant)

colors = [partition[n] for n in G1_giant.nodes()]
my_colors = plt.cm.Set2

plt.figure(figsize=(30,30))
plt.axis('off')

layout = nx.spring_layout(G1_giant)
nx.draw_networkx_nodes(G1_giant, layout, node_color= colors, cmap = my_colors, node_size=100, edgecolors='k', alpha = 0.95)
nx.draw_networkx_edges(G1_giant, layout, edge_color='k',width=1, alpha=0.2)

stats = mk.graphStats(G1_giant)
plt.text(0.8, -0.8, stats, fontsize=20)
plt.title("Citation Network 2014", fontsize=40)

#plt.savefig("Graphs/Citation Network 2014.pdf")
plt.show()
```

# Plotting Holoviews Graphs

The following code creates graphs using Holoviews, which offers an interactive alternative visualization to the Matplotlib graphs.

It is best to save the graphs as external files, since they don't always interact nicely with python notebooks. 

```python
# Set the size of the graphs
%opts Nodes Graph [width=1200 height=1200 xaxis=None yaxis=None]

# Use the renderer to save the files
renderer = hv.renderer('bokeh')
```

### Save the Full Graph

This includes nodes that have no edges or are not part of the biggest cluster. 

```python
RC = RC_ALL
Gx = RC.networkCitation(coreOnly=True)
G = hv.Graph.from_networkx(Gx, nx.spring_layout)
G = G.relabel("SSP Citation Network - Full")
renderer.save(G, 'Graphs/Citation Network Full')

print(mk.graphStats(Gx))
```

### Remove Isolates

Remove any nodes that are not connected to the biggest cluster of the network.

```python
G = RC.networkCitation(coreOnly=True)
mk.dropEdges(G, minWeight = 1, dropSelfLoops = True)
G.remove_nodes_from(list(nx.isolates(G)))

G1 = G.to_undirected()
G1_giant = max(nx.connected_component_subgraphs(G1), key=len)
partition = community.best_partition(G1_giant)

colors = [partition[n] for n in G1_giant.nodes()]
```

```python
G = hv.Graph.from_networkx(G1_giant, nx.spring_layout)
G = G.redim.range(x=(-0.05, 1.05), y=(-0.05, 1.05)).options(cmap='#e5ae38')
G = G.relabel("SSP Citation Network - Simplified")
renderer.save(G, 'Graphs/Citation Network Simplified')

print(mk.graphStats(G1_giant))
```

### Bundle the edges

Using the budle_graph function from Holoviews, the edges are packaged together to allow for a clearer visualization. 

```python
G = bundle_graph(G)
G = G.relabel("SSP Citation Network - Bundled")
renderer.save(G, 'Graphs/Citation Network Bundled')

print(mk.graphStats(G1_giant))
```
